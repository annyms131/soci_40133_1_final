```python
import pandas as pd
import numpy as np
import matplotlib
import seaborn as sns
```

```python
def set_style():
    
    # Set reasonable defaults for font size for figure that will go in a paper
    sns.set_context("paper")
    
    # Set the font to be serif, rather than sans
    sns.set(font='serif')
    
    # Make the background white, and specify the
    # specific font family
    sns.set_style("white", {
        "font.family": "serif",
        "font.serif": ["Times", "Palatino", "serif"]
    })
    
# call the function    
set_style()
```

```python
# define a new year variable for later analysis 
def year (row):
    if  row['YY'] < 1980:
        return "70's"
    elif row['YY'] <1990:
        return "80's"
    elif row['YY'] <2000:
        return "90's"
    elif row['YY'] <2010:
        return "00's"
    elif row['YY'] <2020:
        return "10's"
    else: return "20's"
```

```python
df['year'] = df.apply(lambda x: year(x), axis=1)
df.head()
```
