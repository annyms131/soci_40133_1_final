```python
# define a new year variable for later analysis and save it separately for efficiency
def year (row):
    if  row['YY'] < 1980: # there is one article published in 1968 and will be included here 
        return "70"
    elif row['YY'] < 1990:
        return "80"
    elif row['YY'] < 2000:
        return "90"
    elif row['YY'] < 2010:
        return "00"
    elif row['YY'] < 2020:
        return "10"
    else: return "20"
```

```python
df['year'] = df.apply(lambda x: year(x), axis=1)
df.head()
```

```python
df['year'].value_counts()
```

```python
hist = df['year'].hist()
hist.invert_xaxis()
```

```python
# subset df by `year` and save them separately for later use
df70 = df[df["year"] == "70"]    
df80 = df[df["year"] == "80"]    
df90 = df[df["year"] == "90"]    
df00 = df[df["year"] == "00"]    
df10 = df[df["year"] == "10"]   

df = pd.concat([df70, df80, df90, df00, df10])
```
